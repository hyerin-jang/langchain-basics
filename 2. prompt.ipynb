```python
from langchain_ollama import ChatOllama

llm = ChatOllama(model="llama3.2:1b")

llm.invoke("what is the capital of France?") # prompt: LLM 호출할 때 쓰는 명령어
```

```python
from langchain_core.prompts import PromptTemplate

prompt_template = PromptTemplate(
    template="what is the capital of {country}?",
    input_variables=["country"],
)

prompt = prompt_template.invoke({"country": "France"})

print(prompt)

llm.invoke(prompt_template.invoke({"country": "France"}))
```

```python
from email import message
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage

# few shot: 예제를 넣을수록 정확도가 올라감. 이때 사용하는게 list of messages
messages = [
    SystemMessage(content="You are a helpful assistant"),
    HumanMessage(content="what is the capital of France?"),
    AIMessage(content="The capital of France is Paris."),
    HumanMessage(content="what is the capital of Germany?"),
    AIMessage(content="The capital of Germany is Berlin."),
    HumanMessage(content="what is the capital of Spain?"),
    AIMessage(content="The capital of Spain is Madrid."), 
    HumanMessage(content="what is the capital of {country}?"),
]

llm.invoke(messages) # list 형태로 넣어줘야 함
```

```python
# 조금 더 langchain 스러운 방식
from langchain_core.prompts import ChatPromptTemplate

chat_prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant"),
    ("human", "what is the capital of {country}?"),
])

chat_prompt = chat_prompt_template.invoke({"country": "France"})

print(chat_prompt)

```
